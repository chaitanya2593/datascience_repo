```python
plt.figure(figsize=(20,20))
sns.heatmap(phising_data_frame_corr,annot=True, cmap="YlGnBu")
plt.show()
```

```python
# custom designed function
def extract_x_columns(df_y):
    X = df_y[df_y.between(0.1,0.8)].index        
    X = X.append(df_y[df_y.between(-0.8,-0.1)].index)    
    return X
```

```python
x_columns  = extract_x_columns(phising_data_frame_corr['Result'])
```

```python
X = phising_data_frame[x_columns]
Y = phising_data_frame['Result']
```

### Logistic Regression 

```python
# Train and test split

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.30, random_state=62)
x_train.shape
x_test.shape
y_train.shape
y_test.shape
```

```python
# Building logistic regression model

from sklearn.linear_model import LogisticRegression
from sklearn import metrics
logreg = LogisticRegression(C=0.6)
```

```python
logreg.fit(x_train, y_train)
```

```python
# now applying our learnt model on test and also on train data

y_pred_test = logreg.predict(x_test)
y_pred_train = logreg.predict(x_train)
```

```python
# comparing the metrics of predicted lebel and real label of test data
print("Test Accuracy: ", metrics.accuracy_score(y_test, y_pred_test))
```

```python
# comparing the metrics of predicted lebel and real label of test data
print("Train Accuracy: ", metrics.accuracy_score(y_train, y_pred_train))
```

```python
# creating a confusion matrix to understand the classification
conf = metrics.confusion_matrix(y_test, y_pred_test)
```

```python
print(conf)
```

```python
cmap = sns.cubehelix_palette(50, hue=0.05, rot=0, light=0.9, dark=0, as_cmap=True)
sns.heatmap(conf,cmap = cmap,xticklabels=['Prediction No','Prediction Yes'],yticklabels=['Actual No','Actual Yes'], annot=True,
            fmt='d')
```

```python
# Creating Classification Report

cr = metrics.classification_report(y_test, y_pred_test)
print(cr)
```

##  KNN

```python
from sklearn.model_selection import cross_val_score
```

```python
X = phising_data_frame.drop('Result', axis=1)
Y = phising_data_frame['Result']
```

```python
# Train and test split

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.30, random_state=62)
x_train.shape
x_test.shape
y_train.shape
y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors=7)
```

```python
KNN.fit(x_train, y_train)
```

```python
predict_y_knn = KNN.predict(x_test)
```

```python
# Creating Classification Report

cr = metrics.classification_report(y_test, predict_y_knn)
print(cr)
```

```python
# comparing the metrics of predicted lebel and real label of test data
print("Test Accuracy: ", metrics.accuracy_score(y_test, predict_y_knn))
```

```python
# creating a confusion matrix to understand the classification
conf = metrics.confusion_matrix(y_test, predict_y_knn)
```

```python
print(conf)
```

```python
cmap = sns.cubehelix_palette(50, hue=0.05, rot=0, light=0.9, dark=0, as_cmap=True)
sns.heatmap(conf,cmap = cmap,xticklabels=['Prediction No','Prediction Yes'],yticklabels=['Actual No','Actual Yes'], annot=True,
            fmt='d')
```

```python
KNN_string = [1,3,5,7,9,11,13,15,17]
for c in KNN_string:
    KNN = KNeighborsClassifier(n_neighbors=c)
    KNN.fit(x_train, y_train)
    predict_y_knn = KNN.predict(x_test)
    print("Test Accuracy when n_neighbors is "  + str(c) + ": " , metrics.accuracy_score(y_test, predict_y_knn))
    score_knn = cross_val_score(KNN, x_test, y_test, cv=5)
    print("score_knn" + str(score_knn))
    print ("Mean Score KNN " + str(score_knn.mean()))
```

```python

KNN = KNeighborsClassifier(n_neighbors=13)
KNN.fit(x_train, y_train)
predict_y_knn = KNN.predict(x_test)
print("Test Accuracy when n_neighbors is 13: " , metrics.accuracy_score(y_test, predict_y_knn))
score_knn = cross_val_score(KNN, x_test, y_test, cv=5)
score_knn.mean()
```

```python
score_lr = cross_val_score(logreg, x_test, y_test, cv=5)
print("\nscore_dtc" + str(score_lr))
print ("Mean Score Logistic Regression :: " + str(score_lr.mean()))


score_dtc = cross_val_score(DTC, x_test, y_test, cv=5)
print("\nscore_dtc" + str(score_dtc))
print ("Mean Score Decision Tree :: " + str(score_dtc.mean()))


score_knn = cross_val_score(KNN, x_test, y_test, cv=5)
print("\nscore_knn" + str(score_knn))
print ("Mean Score KNN :: " + str(score_knn.mean()))

```
